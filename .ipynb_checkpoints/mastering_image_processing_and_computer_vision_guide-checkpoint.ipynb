```python
import os
import sys
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
print('OpenCV-Python Lib Version:', cv2.__version__)
print('Python Version:',sys.version)
```

```python
path_to_img = 'rgbimg/fruit-bowl.jpg'
```

## Understand Basic Image Handling: 
+ Read the given image using OpenCV-Python functions.
+ Find out Image size, Image datatype and Pixel value at specific location of Image.

```python
MainImgBGR = cv2.imread(path_to_img)
MainImgGray = cv2.imread(path_to_img,cv2.IMREAD_GRAYSCALE)

print('Datatype Class of Image:',type(MainImgBGR))

#For Color Image
print('[B G R] Image Shape:',MainImgBGR.shape)
print('[GrayScale] Image Shape:',MainImgBGR.shape)

print('[B G R] Values at location(100,200):' ,MainImgBGR[100,200])
print('[GrayScale] Values at location(100,200):' ,MainImgGray[100,200])

#For GrayImagr

plt.subplot(1,2,1)
plt.imshow(MainImgGray)
plt.subplot(1,2,2)
plt.imshow(MainImgBGR)
plt.show()
```

```python
MainImgRGB = cv2.cvtColor(MainImgBGR,cv2.COLOR_BGR2RGB)

plt.title('Main-Original Image',fontweight='bold')
plt.imshow(MainImgRGB)
```

```python
plt.subplots_adjust(wspace=0.5)
plt.subplot(1,3,1)
plt.title('Image Channel: R')
plt.imshow(MainImgRGB[:,:,0])
plt.subplot(1,3,2)
plt.title('Image Channel: G')
plt.imshow(MainImgRGB[:,:,1])
plt.subplot(1,3,3)
plt.title('Image Channel: B')
plt.imshow(MainImgRGB[:,:,2])
plt.show()
```
